### Linear Regression

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("housing.csv")
df
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
X = np.array(df.iloc[:,1])
y = np.array(df.iloc[:,0])
X = X.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
model = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state = 1)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
r2 = r2_score(y_test,y_pred)
print(r2)
```

```python
plt.scatter(X,y, color='blue', label='Actual data')
plt.plot(X_test,y_pred,color='red', linewidth=2, label= 'Best Fit Line')
```

```python
slope = model.coef_[0]
intercept = model.intercept_
print("slope : ",slope)
print("intercept : ",intercept)
```

#### advantages of the linear regression
- if data is increasing linearly then this model predicts with high accuracy

#### Disadvantages
- if data is not dependent on single variable then model can give worst accuracy
- extremely bad prediction for outliers
